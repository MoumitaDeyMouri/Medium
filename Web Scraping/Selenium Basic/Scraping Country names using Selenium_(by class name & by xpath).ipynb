## Import Modules 

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import pandas as pd
```

Set path from Chrome Driver

```python
path = "C:/Users/ACER/Desktop/chromedriver.exe"
```

```python
# setting path for the driver
browser = webdriver.Chrome(executable_path = path)
```

```python
url = "https://www.scrapethissite.com/pages/simple/"
```

```python
# open the page url in chrome
browser.get(url)
```

## Scrape the data

```python
# get country name
countries = browser.find_elements_by_xpath("//h3[@class='country-name']")
countries                                  
```

```python
# parse the data
country_list = []
for i in countries:
    #get the text data
    temp = i.text
    country_list.append(temp)
    
country_list
```

```python
# get the country population
populations = browser.find_elements_by_class_name('country-population')
populations
```

```python
# parse the population data
population_list = []
for i in populations:
    #get the text data
    temp = i.text
    population_list.append(temp)
    
population_list
```

```python
# get the capital of the country
capitals = browser.find_elements_by_xpath("//span[@class='country-capital']")
capitals
```

```python
# parse the capitals data
capital_list = []

for i in capitals:
    # get the text data
    temp = i.text
    capital_list.append(temp)
    
capital_list
```

```python
# get the country area(km**2)
areas = browser.find_elements_by_class_name('country-area')
areas
```

```python
# parse the area data of countries
area_list = []

for i in areas:
    # get the text data
    temp = i.text
    area_list.append(temp)
area_list
```

```python
print(len(country_list))
print(len(capital_list))
print(len(population_list))
print(len(area_list))
```

### Storing the Scraped data

```python
df = pd.DataFrame()
df['Country Name'] = country_list
df['Capital'] = capital_list
df['Population'] = population_list
df['Area (km**2)'] = area_list

df.head(5)
```

### -- Store the data & close the browser.

```python
df.to_csv('countries.csv', index=False)
# or as xlsx format
df.to_excel('countries.xlsx', index=False)

browser.quit()
```
